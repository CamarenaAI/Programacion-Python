## Interpretación
Las medias y totales mensuales son en base a los datos disponibles, cuando en las medias aparece algún resultado en rojo, significa que no se dispone de información del mes completo, en este caso, la media o total es de los días de los que existen datos.

- T:	Temperatura media (°C)
- TM:	Temperatura máxima (°C)
- Tm:	Temperatura mínima (°C)
- SLP:	Presión atmosférica a nivel del mar (hPa)
- H:	Humedad relativa media (%)
- PP:	Precipitación total de lluvia y/o nieve derretida (mm)
- VV:	Visibilidad media (Km)
- V:	Velocidad media del viento (Km/h)
- VM:	Velocidad máxima sostenida del viento (Km/h)
- VG:	Velocidad de ráfagas máximas de viento (Km/h)
- RA:	Indica si hubo lluvia o llovizna (En la media mensual, total días que llovió)
- SN:	Indicador de nieve (En la media mensual, total días que nevó)
- TS:	Indica si hubo tormenta (En la media mensual, total días con tormenta)
- FG:	Indica si hubo niebla (En la media mensual, total días con niebla)


```python
# Importar pandas
import pandas as pd

# Cargar el archivo CSV en un DataFrame
datos = pd.read_csv('datos.csv')
datos
```

Al tener valores nulos en algunas columnas y filas se realizo una limpieza de datos para poder trabajar con los datos de una manera mas eficaz

```python
# Eliminar las columnas con valores NaN
datos = datos.drop('PP', axis=1)
datos = datos.drop('VV', axis=1)
datos = datos.drop('V', axis=1)
datos = datos.drop('VM', axis=1)
datos = datos.drop('VG', axis=1)
datos = datos.drop('RA', axis=1)
datos = datos.drop('SN', axis=1)
datos = datos.drop('TS', axis=1)
datos = datos.drop('FG', axis=1)

# Eliminar las filas con valores NaN
datos = datos.drop(20)
datos = datos.drop(26)
datos = datos.drop(27)

datos
```

Despues de realizar nuestra limpieza de datos se muestra la temperatura media, temperatura mínima y temperatura 
máxima de la ciudad de Guadalajara, Jalisco, así como la presión atmosférica y la humedad relativa.<br>
La temperatura mínima y máxima entra entre los parámetros ya que la temperatura más baja 
registrada en todo el año es de 5°C y la temperatura mas alta es de 32°C.<br>
